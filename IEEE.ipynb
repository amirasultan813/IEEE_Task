### Install fuzzywazzy to use it for string matching

```python
%pip install fuzzywuzzy   
```

```python
from fuzzywuzzy import process
import warnings
warnings.filterwarnings("ignore")
```

```python
categoryFcountry=['USA','United States','Canada','America','UK','United Kingdom']
```

```python
for stat in categoryFcountry:
    matches = process.extract(stat,data['Country'], limit=data.shape[0])
    for potential_match in matches:
        if potential_match[1] >= 70:
            data.loc[data['Country'] == potential_match[0], 'Country'] = stat
data['Country'].value_counts()
```

```python
data['Country'].unique()
```

```python
for x in data['Country']:
    if ((('U' in str(x) and 'S' in str (x)))):
        data['Country']=data['Country'].replace(x,"USA")
```

```python
data['Country']=data['Country'].replace(['United States','America','New York','MEXICO','CALIFORNIA','NEW JERSEY','U.S.A.','US OF A','U S A'],'USA')
data['Country']=data['Country'].replace(['United Kingdom','ENGLAND','ENDLAND'],'UK')
```

```python
data['Country'].unique()
```

```python
data['Country'].value_counts()
```

```python
remain= [x for x in data['Country'].unique()]
remain.remove('USA')
remain.remove('Canada')
remain.remove('UK')
remain.remove('GERMANY')
remain.remove('NETHERLANDS')
remain.remove('JAPAN')
remain.remove('SCOTLAND')
remain.remove('SWITZERLAND')
remain.remove('IRELAND')
print(remain)
```

```python
data['Country']=data['Country'].replace(remain,'Other')
```
